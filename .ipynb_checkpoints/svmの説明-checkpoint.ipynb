```python
import numpy as np

x1 = np.linspace(0.0, 10.0, 100)
x2 = np.linspace(0.0, 10.0, 100)
print(x1)
print(x1.shape)

print('-------------------')

print(x2)

X1, X2 = np.meshgrid(x1, x2)   
print(X1)
print(X1.shape)

print('---------------------')

print(X2)
print(X2.shape)
```

```python
print(X1.ravel().shape)
print(X2.ravel().shape)

plot_X = np.c_[X1.ravel(), X2.ravel()]
print(plot_X.shape)
```

```python
print(plot_X)

print(plot_X.T)
```

```python
print(plot_X.T[0])
```
