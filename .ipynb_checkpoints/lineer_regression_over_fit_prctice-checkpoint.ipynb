```python
from sklearn import datasets 
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#dataには患者のバイタル値。targetには25～346の値で表された一年後の進行状況が格納されている。
dataset = datasets.load_diabetes()
print(dataset.data.shape)
print(dataset.target.shape)
dataset.data
```

```python
#散布図にて可視化
x = dataset.data
y = dataset.target

for i in range(x.shape[1]):
    plt.plot(x[:,i],y,'o')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#ヒストグラム
plt.hist(x)

plt.grid(True)
plt.show()
```

```python
#ヒートマップ
sns.set()
corr = np.corrcoef(x.T,y)
print(corr)
print(corr.shape)
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#オーバーフィッティング対策なしに普通に重回帰分析

x_train,x_test, y_train, y_test = sklearn.model_selection.train_test_split(x,y,test_size=0.5,random_state=0)

clf = linear_model.LinearRegression()

clf.fit(x_train,y_train)

a = clf.coef_
b = clf.intercept_

print('回帰係数：', a)
print('切片：', b)
print('決定係数(学習用）：', clf.score(x_train, y_train))
print('決定係数(テスト用):', clf.score(x_test, y_test))
```

```python
#相関関係の強い説明変数を削除してオーバーフィッティング対策

dataset = datasets.load_diabetes()

x = dataset.data
y = dataset.target

x = np.hstack((x[:,0:5], x[:,8:10]))
              
x_train,x_test, y_train,y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.5, random_state=0)

clf = linear_model.LinearRegression()
clf.fit(x_train,y_train)

a = clf.coef_
b = clf.intercept_

print('回帰係数：', a)
print('切片：', b)
print('決定係数(学習用）：', clf.score(x_train, y_train))
print('決定係数(テスト用):', clf.score(x_test, y_test))
```

```python
#L1正則化をした線形回帰モデルを使いオーバーフィッティング対策

dataset = datasets.load_diabetes()

x = dataset.data
y = dataset.target

x_train,x_test, y_train,y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.5, random_state=0)

clf = linear_model.Lasso()
clf.fit(x_train,y_train)

a = clf.coef_
b = clf.intercept_

print('回帰係数：', a)
print('切片：', b)
print('決定係数(学習用）：', clf.score(x_train, y_train))
print('決定係数(テスト用):', clf.score(x_test, y_test))
```

```python
#L2正則化をした線形回帰モデルを使いオーバーフィッティング対策

dataset = datasets.load_diabetes()

x = dataset.data
y = dataset.target

x_train,x_test, y_train,y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.5, random_state=0)

clf = linear_model.Ridge()
clf.fit(x_train,y_train)

a = clf.coef_
b = clf.intercept_

print('回帰係数：', a)
print('切片：', b)
print('決定係数(学習用）：', clf.score(x_train, y_train))
print('決定係数(テスト用):', clf.score(x_test, y_test))
```
